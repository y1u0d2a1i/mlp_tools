```python
import os
import numpy as np

with open(os.path.join(path2input, "si.com"), "r") as f:
    ga_input_lines = [s.strip() for s in f.readlines()]

distance_list = np.linspace(0.1, 6.5, 50)

def get_changed_ga_line(line: str, distance) -> str:
    splitted_line = line.split()
    splitted_line[-1] = str(distance)
    return " ".join(splitted_line)

for distance in distance_list:
    path2target = os.path.join(path2root, f"ga_{distance:.2f}")
    if not os.path.exists(path2target):
        os.makedirs(path2target)
    
    copied_ga_input_lines = ga_input_lines.copy()
    change_atom_line = get_changed_ga_line(
        line=copied_ga_input_lines[-2],
        distance=distance
    )
    copied_ga_input_lines[-2] = change_atom_line
    copied_ga_input_lines.append("")

    with open(os.path.join(path2target, "si.com"), "w") as f:
        f.write("\n".join(copied_ga_input_lines))
```

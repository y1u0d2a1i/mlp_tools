# Pair Potentialのプロット

## Quantum espresso

```python
from glob import glob
import pandas as pd
import matplotlib.pyplot as plt
from mlptools.io.read import read_from_format
from mlptools.analyzer.pair_potential import PairPotentialAnalyzer
```

### Si-O

```python
# Si-O
path2sio = "/Users/y1u0d2/desktop/Lab/result/qe/SiO2/mp-dimer/Si-O/result"
atoms_dirs = glob(f"{path2sio}/scf*")
print(f"Number of dirs {len(atoms_dirs)}")

analyzer = PairPotentialAnalyzer()
pair_potential_df_sio = analyzer.get_from_espresso(atoms_dirs)

# plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(pair_potential_df_sio["distance"], pair_potential_df_sio["energy"], marker="o")
ax.set_xlabel("Distance (Angstrom)")
ax.set_ylabel("Energy (eV)")
ax.set_title("Two-body potential of Si-O")
```

```python
# with spin
path2sio = "/Users/y1u0d2/desktop/Lab/result/qe/SiO2/mp-dimer/Si-O/spin/result"
atoms_dirs = glob(f"{path2sio}/scf*")
print(f"Number of dirs {len(atoms_dirs)}")

analyzer = PairPotentialAnalyzer()
pair_potential_df_sio_spin = analyzer.get_from_espresso_mag(atoms_dirs)

# plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(pair_potential_df_sio_spin["distance"], pair_potential_df_sio_spin["energy"], marker="o")
ax.set_xlabel("Distance (Angstrom)")
ax.set_ylabel("Energy (eV)")
ax.set_title("Two-body potential of Si-O")
```

```python
# plot together
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(pair_potential_df_sio["distance"], pair_potential_df_sio["energy"], marker="o", label="no spin")
ax.plot(pair_potential_df_sio_spin["distance"], pair_potential_df_sio_spin["energy"], marker="o", label="spin")
ax.set_xlabel("Distance (Angstrom)")
ax.set_ylabel("Energy (eV)")
ax.set_title("Two-body potential of Si-O")
ax.legend()
```

```python
import pandas as pd
import numpy as np
from glob import glob

path2data = "/Users/y1u0d2/desktop/Lab/result/lammps/Si/sputtering/ml/model_28/0deg_750eV"
```

```python
data_files = glob(f"{path2data}/data.*")
data_files = sorted(data_files, key=lambda x: int(x.split(".")[-1]))
len(data_files)
```

```python
timestep_list = []
n_atoms_list = []

for i, data_file in enumerate(data_files):
    with open(data_file) as f:
        lines = [s.strip() for s in f.readlines()]
    for l in lines:
        if "atoms" in l:
            n_atoms = int(l.split()[0])
            break
    timestep_list.append(i*20000)
    n_atoms_list.append(n_atoms)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(timestep_list, n_atoms_list)
ax.set_xlabel("timestep")
ax.set_ylabel("n_atoms")
ax.set_title("n_atoms vs. timestep")
plt.show()
```

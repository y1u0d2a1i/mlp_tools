```python
from glob import glob
import pandas as pd
import matplotlib.pyplot as plt
from mlptools.utils.constants import ZERO_POINT_ENERGY_OXYGEN_ESPRESSO, ZERO_POINT_ENERGY_OXYGEN_GAUSSIAN, ZERO_POINT_ENERGY_SILICON_GAUSSIAN, ZERO_POINT_ENERGY_SILICON_ESPRESSO

path2pairpotential = "/Users/y1u0d2/desktop/Lab/result/pair_potential"
```

# Si-Si

```python
import plotly.graph_objects as go

# load csv
espresso_df = pd.read_csv(path2pairpotential+"/espresso_Si-Si.csv")
gaussian_df = pd.read_csv(path2pairpotential+"/gaussian_Si-Si.csv")
matlantis_df = pd.read_csv(path2pairpotential+"/matlantis_Si-Si.csv")

# plot
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
ax.plot(espresso_df["distance"], espresso_df["energy"] - 2 * ZERO_POINT_ENERGY_SILICON_ESPRESSO, label="Quantum Espresso", marker="o")
ax.plot(gaussian_df["distance"], gaussian_df["energy"] - 2 * ZERO_POINT_ENERGY_SILICON_GAUSSIAN, label="Gaussian", marker="o")
# ax.plot(matlantis_df["distance"], matlantis_df["energy"], label="Matlantis", marker="o")
ax.set_title("Pair Potential of Si-Si")
ax.set_xlabel("distance (Angstrom)")
ax.set_ylabel("energy (eV)")
ax.set_xlim(0, 5.0)
ax.set_ylim(-20, 100)
ax.legend()
```

```python
# plot with plotly
fig = go.Figure()
fig.add_trace(go.Scatter(x=espresso_df["distance"], y=espresso_df["energy"] - 2 * ZERO_POINT_ENERGY_SILICON_ESPRESSO, name="Quantum Espresso"))
fig.add_trace(go.Scatter(x=gaussian_df["distance"], y=gaussian_df["energy"] - 2 * ZERO_POINT_ENERGY_SILICON_GAUSSIAN, name="Gaussian"))
fig.add_trace(go.Scatter(x=matlantis_df["distance"], y=matlantis_df["energy"], name="Matlantis"))
fig.update_layout(
    title="Pair Potential of Silicon",
    xaxis_title="distance (Angstrom)",
    yaxis_title="energy (eV)",
    font=dict(
        size=18,
        color="RebeccaPurple"
    )
)
fig.show()
```

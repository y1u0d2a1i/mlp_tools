## LAMMPSのdumpファイルから入射直後の構造を取得する

```python
path2dump = '/Users/y1u0d2/desktop/Lab/result/lammps/Si/sputtering/ml/depth15/0deg_1000eV/dump.lammpstrj'
# Import a sequence of files.
pipeline = import_file(path2dump)
```

```python
def get_ase_atoms_from_lmp_dump_by_frame_index(pipeline, frame_index):
    data = pipeline.source.compute(frame_index)
    ase_atoms = ovito_to_ase(data)
    print(f'Successfully get ase atoms from frame {frame_index}')
    print(f'Number of atoms: {len(ase_atoms)}')
    return ase_atoms
```

```python
dump_structure_every_n_steps = 500
insert_atom_every_n_steps = 20000

insert_atom_every_n_frames = insert_atom_every_n_steps // dump_structure_every_n_steps
pipeline.source.num_frames, insert_atom_every_n_frames

insert_atom_frame_idx_list = [i * insert_atom_every_n_frames + 1 for i in range(pipeline.source.num_frames // insert_atom_every_n_frames)]
```

```python
# get random 5 value from list
frame_idx_list = []
for idx in random.sample(insert_atom_frame_idx_list, 10):
    n_frame_before_after = 5
    frame_idx_list_before_after = [idx - i for i in range(n_frame_before_after, 0, -1)] + [idx + i for i in range(1, n_frame_before_after + 1)]
    frame_idx_list.extend(frame_idx_list_before_after)
```

```python
frame_idx_list = sorted(list(set(frame_idx_list)))
print(f'number of frames: {len(frame_idx_list)}')
```

```python
path2save = '/Users/y1u0d2/desktop/Lab/result/lammps/Si/sputtering/ml/depth15/0deg_1000eV/atoms_near_injection'
for frame_idx in frame_idx_list:
    filename = f'frame_{frame_idx}.pkl'
    ase_atoms = get_ase_atoms_from_lmp_dump_by_frame_index(pipeline, frame_idx)

    with open(os.path.join(path2save, filename), 'wb') as f:
        pickle.dump(ase_atoms, f)
```

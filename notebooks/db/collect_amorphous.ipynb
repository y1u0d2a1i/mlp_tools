```python
from glob import glob
import os
from mlptools.io.read import read_from_format
```

```python
amorphous_files = glob("/Users/y1u0d2/desktop/Lab/result/qe/Si/amorphous/**/scf.out", recursive=True)
```

```python
all_mlpatoms = []
for amorphous_file in amorphous_files:
    path2dir = os.path.dirname(amorphous_file)
    try:
        mlpatoms = read_from_format(
            path2target=path2dir,
            format="espresso-in"
        )
        all_mlpatoms.append(mlpatoms)
    except:
        print(path2dir)
        continue
```

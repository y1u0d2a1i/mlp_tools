# フォルダ内にある使用していないフォルダを削除する

```python
from sqlalchemy import create_engine, Column, Integer, String, Sequence
from sqlalchemy.orm import declarative_base
import uuid
from sqlalchemy.orm import sessionmaker
import os
import shutil
from mlptools.io.read import read_from_format
import pickle

path2db = "/Users/y1u0d2/desktop/Lab/data/qe_data/Si"
path2data = "/Users/y1u0d2/desktop/Lab/data/qe_data/Si/atoms"

Base = declarative_base()

class Structure(Base):
    __tablename__ = 'structure'
    id = Column(String, primary_key=True)
    original_path = Column(String, unique=True)
    structure_id = Column(Integer)
    structure_name = Column(String)
    calculation_type = Column(String)

# データベースエンジンの作成とテーブルの初期化
engine = create_engine(f'sqlite:///{path2db}/structure.db')
Base.metadata.create_all(engine)
```

```python
session = sessionmaker(bind=engine)()
all_used_uuid = session.query(Structure.id).all()
all_used_uuid = [uuid[0] for uuid in all_used_uuid]

all_atoms_dir = os.listdir(path2data)
```

```python
print(f"Number of uuid in db: {len(all_used_uuid)}")
print(f"Number of atoms dir: {len(all_atoms_dir)}")
```

```python
# get difference between all_atoms_dir and all_used_uuid using set
remove_uuid_list = list(set(all_atoms_dir) - set(all_used_uuid))
print(f"Number of uuid to be removed: {len(remove_uuid_list)}")
```

```python
# remove directory and files
for uuid in remove_uuid_list:
    shutil.rmtree(f"{path2data}/{uuid}")
    print(f"Remove: {uuid}")
```
